```python
import pandas as pd
df.columns = df.columns.str.strip()

# Convert everything to string
df["promotion_flag"] = df["promotion_flag"].astype(str)

# Strip extra spaces and lowercase all values
df["promotion_flag"] = df["promotion_flag"].str.strip().str.lower()

# Replace possible text forms with numeric 0/1
df["promotion_flag"] = df["promotion_flag"].replace({
    "yes": 1, "y": 1, "true": 1, "t": 1, "1": 1,
    "no": 0, "n": 0, "false": 0, "f": 0, "0": 0
})

# Convert safely to numeric
df["promotion_flag"] = pd.to_numeric(df["promotion_flag"], errors="coerce").fillna(0).astype(int)

# Convert to True/False (boolean)
df["promotion_flag"] = df["promotion_flag"].astype(bool)

# Check result
print(" promotion_flag cleaned successfully.")
print(df["promotion_flag"].value_counts())
```
